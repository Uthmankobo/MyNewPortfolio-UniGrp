```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the scaled 1,650 MW plant
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
energy_kwh = [145140, 139640, 154180, 147810, 143410, 130150, 123710, 115740, 125650, 141240, 140270, 142840]
energy_mwh = [145.14, 139.64, 154.18, 147.81, 143.41, 130.15, 123.71, 115.74, 125.65, 141.24, 140.27, 142.84]

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Energy in kWh
ax1.plot(months, energy_kwh, marker='o', linewidth=2, markersize=6, color='#2E86AB', label='Monthly Output')
ax1.set_title('Kolmani Field - Scaled Monthly Energy Output\n(1,650 MW Plant) - kWh', fontsize=14, fontweight='bold')
ax1.set_xlabel('Month', fontsize=12)
ax1.set_ylabel('Energy Output (kWh)', fontsize=12)
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# Add value annotations on the kWh plot
for i, v in enumerate(energy_kwh):
    ax1.annotate(f'{v:,}', (i, v), textcoords="offset points", xytext=(0,10),
                ha='center', fontsize=8, alpha=0.7)

# Plot 2: Energy in MWh
ax2.plot(months, energy_mwh, marker='s', linewidth=2, markersize=6, color='#A23B72', label='Monthly Output')
ax2.set_title('Kolmani Field - Scaled Monthly Energy Output\n(1,650 MW Plant) - MWh', fontsize=14, fontweight='bold')
ax2.set_xlabel('Month', fontsize=12)
ax2.set_ylabel('Energy Output (MWh)', fontsize=12)
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='x', rotation=45)

# Add value annotations on the MWh plot
for i, v in enumerate(energy_mwh):
    ax2.annotate(f'{v}', (i, v), textcoords="offset points", xytext=(0,10),
                ha='center', fontsize=8, alpha=0.7)

plt.tight_layout()
plt.show()

# Additional single detailed plot
plt.figure(figsize=(12, 6))
plt.plot(months, energy_kwh, marker='o', linewidth=2.5, markersize=8,
         color='#FF6B6B', label='Energy Output')

plt.title('Kolmani Oil Field - Photovoltaic Energy Production (Scaled to 1,650 MW)\n',
          fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Energy Output (kWh)', fontsize=12)
plt.grid(True, alpha=0.3)

# Add detailed value annotations
for i, v in enumerate(energy_kwh):
    plt.annotate(f'{v:,} kWh', (i, v), textcoords="offset points",
                xytext=(0,15), ha='center', fontsize=9,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

# Highlight peak and lowest months
peak_idx = energy_kwh.index(max(energy_kwh))
low_idx = energy_kwh.index(min(energy_kwh))

plt.annotate('Peak: March', (peak_idx, energy_kwh[peak_idx]),
             textcoords="offset points", xytext=(0,30),
             ha='center', fontsize=10, fontweight='bold',
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.8))

plt.annotate('Lowest: August', (low_idx, energy_kwh[low_idx]),
             textcoords="offset points", xytext=(0,-40),
             ha='center', fontsize=10, fontweight='bold',
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral", alpha=0.8))

plt.tight_layout()
plt.show()

# Print summary statistics
print("="*60)
print("KOLMANI FIELD SOLAR PRODUCTION SUMMARY (Scaled to 1,650 MW)")
print("="*60)
print(f"Total Annual Production: {sum(energy_kwh):,} kWh")
print(f"Total Annual Production: {sum(energy_mwh):.2f} MWh")
print(f"Peak Month: {months[peak_idx]} ({energy_kwh[peak_idx]:,} kWh)")
print(f"Lowest Month: {months[low_idx]} ({energy_kwh[low_idx]:,} kWh)")
print(f"Monthly Average: {np.mean(energy_kwh):,.0f} kWh")
print(f"Variability (Range): {max(energy_kwh)-min(energy_kwh):,.0f} kWh")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the scaled 1,650 MW plant
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']
months_short = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

energy_kwh = [145140, 139640, 154180, 147810, 143410, 130150,
              123710, 115740, 125650, 141240, 140270, 142840]
energy_mwh = [145.14, 139.64, 154.18, 147.81, 143.41, 130.15,
              123.71, 115.74, 125.65, 141.24, 140.27, 142.84]

# Color scheme based on energy production (green for high, red for low)
colors = []
for energy in energy_kwh:
    if energy >= 150000:
        colors.append('#2E8B57')  # High production - green
    elif energy >= 140000:
        colors.append('#7FB3D5')  # Medium-high - blue
    elif energy >= 130000:
        colors.append('#F7DC6F')  # Medium - yellow
    elif energy >= 120000:
        colors.append('#F39C12')  # Medium-low - orange
    else:
        colors.append('#E74C3C')  # Low production - red

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Plot 1: Bar chart in kWh
bars1 = ax1.bar(months_short, energy_kwh, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
ax1.set_title('Kolmani Field - Monthly Photovoltaic Energy Production\n(Scaled to 1,650 MW Plant) - kWh',
              fontsize=16, fontweight='bold', pad=20)
ax1.set_ylabel('Energy Output (kWh)', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')
ax1.tick_params(axis='x', rotation=45)

# Add value labels on bars for kWh
for bar, value in zip(bars1, energy_kwh):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 2000,
             f'{value:,} kWh', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Plot 2: Bar chart in MWh
bars2 = ax2.bar(months_short, energy_mwh, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)
ax2.set_title('Kolmani Field - Monthly Photovoltaic Energy Production\n(Scaled to 1,650 MW Plant) - MWh',
              fontsize=16, fontweight='bold', pad=20)
ax2.set_ylabel('Energy Output (MWh)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Month', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')
ax2.tick_params(axis='x', rotation=45)

# Add value labels on bars for MWh
for bar, value in zip(bars2, energy_mwh):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 2,
             f'{value} MWh', ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# Single detailed bar chart with seasonal grouping
plt.figure(figsize=(16, 8))

# Create the bar chart
bars = plt.bar(months, energy_kwh, color=colors, alpha=0.8, edgecolor='black', linewidth=0.8)

plt.title('Kolmani Oil Field - Detailed Monthly Energy Production Analysis\n(Scaled to 1,650 MW Photovoltaic Plant)\n',
          fontsize=18, fontweight='bold')
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Energy Output (kWh)', fontsize=14, fontweight='bold')

# Add value labels on bars
for bar, value in zip(bars, energy_kwh):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1000,
             f'{value:,} kWh', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Add seasonal annotations
plt.axvline(x=4.5, color='red', linestyle='--', alpha=0.7, linewidth=1)
plt.axvline(x=8.5, color='blue', linestyle='--', alpha=0.7, linewidth=1)

plt.text(2, max(energy_kwh) * 0.9, 'DRY SEASON\n(High Production)',
         ha='center', va='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))

plt.text(6.5, max(energy_kwh) * 0.9, 'RAINY SEASON\n(Low Production)',
         ha='center', va='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral", alpha=0.7))

plt.text(10.5, max(energy_kwh) * 0.9, 'RECOVERY PERIOD\n(Medium Production)',
         ha='center', va='center', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightyellow", alpha=0.7))

plt.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Horizontal bar chart for better comparison
plt.figure(figsize=(14, 10))
bars_horizontal = plt.barh(months, energy_kwh, color=colors, alpha=0.8, edgecolor='black', linewidth=0.5)

plt.title('Kolmani Field - Monthly Energy Production Comparison\n(Scaled to 1,650 MW Plant)\n',
          fontsize=16, fontweight='bold')
plt.xlabel('Energy Output (kWh)', fontsize=12, fontweight='bold')

# Add value labels on horizontal bars
for bar, value in zip(bars_horizontal, energy_kwh):
    width = bar.get_width()
    plt.text(width + 2000, bar.get_y() + bar.get_height()/2.,
             f'{value:,} kWh', ha='left', va='center', fontsize=10, fontweight='bold')

plt.grid(True, alpha=0.3, axis='x')
plt.tight_layout()
plt.show()

# Print comprehensive summary
print("="*70)
print("KOLMANI FIELD SOLAR PRODUCTION ANALYSIS (Scaled to 1,650 MW)")
print("="*70)

# Calculate statistics
peak_production = max(energy_kwh)
lowest_production = min(energy_kwh)
peak_month = months[energy_kwh.index(peak_production)]
lowest_month = months[energy_kwh.index(lowest_production)]
average_production = np.mean(energy_kwh)

print(f"\n PRODUCTION STATISTICS:")
print(f"    Total Annual Production: {sum(energy_kwh):,} kWh")
print(f"    Total Annual Production: {sum(energy_mwh):.2f} MWh")
print(f"    Peak Production: {peak_month} - {peak_production:,} kWh")
print(f"    Lowest Production: {lowest_month} - {lowest_production:,} kWh")
print(f"    Monthly Average: {average_production:,.0f} kWh")
print(f"    Production Range: {peak_production - lowest_production:,} kWh")

print(f"\n  SEASONAL ANALYSIS:")
dry_season = energy_kwh[0:5]  # Jan-May
rainy_season = energy_kwh[5:9]  # Jun-Sep
recovery_season = energy_kwh[9:12]  # Oct-Dec

print(f"    Dry Season (Jan-May) Average: {np.mean(dry_season):,.0f} kWh")
print(f"    Rainy Season (Jun-Sep) Average: {np.mean(rainy_season):,.0f} kWh")
print(f"    Recovery Period (Oct-Dec) Average: {np.mean(recovery_season):,.0f} kWh")
print(f"    Seasonal Variation: {(max(dry_season) - min(rainy_season)) / max(dry_season) * 100:.1f}%")

print(f"\n COLOR LEGEND:")
print("     Dark Green: Very High Production (>150,000 kWh)")
print("     Blue: High Production (140,000-150,000 kWh)")
print("     Yellow: Medium Production (130,000-140,000 kWh)")
print("     Orange: Low-Medium Production (120,000-130,000 kWh)")
print("     Red: Low Production (<120,000 kWh)")
```
