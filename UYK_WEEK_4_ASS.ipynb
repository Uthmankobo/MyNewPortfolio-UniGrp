```python
import pandas as pd
from IPython.display import display
import nltk
import os

# Set up a custom nltk data folder
nltk_data_path = '/content/nltk_data'
os.makedirs(nltk_data_path, exist_ok=True)

# Add the custom data path to NLTK
nltk.data.path.append(nltk_data_path)

# Download the 'punkt' resource (already done, but good to keep for completeness)
nltk.download('punkt', download_dir=nltk_data_path, quiet=True)

# Download the 'punkt_tab' resource to the same custom directory
nltk.download('punkt_tab', download_dir=nltk_data_path, quiet=True)


# Load the CSV file, ensuring the file path is a string
df = pd.read_csv('quotes_dataset.csv')

# Display the first few rows
display(df.head())

# Whitespace Tokenizer - Splits by spaces only
def whitespace_tokenizer(text):
    return text.split()

# Apply to the dataset
df['Whitespace_Tokens'] = df['Sentence'].apply(whitespace_tokenizer)

# Preview
display(df[['Sentence', 'Whitespace_Tokens']].head())

import re

# Regex Tokenizer - Extracts only words (removes punctuation)
def regex_tokenizer(text):
    return re.findall(r'\b\w+\b', text)

# Apply to the dataset
df['Regex_Tokens'] = df['Sentence'].apply(regex_tokenizer)

# Preview
display(df[['Sentence', 'Regex_Tokens']].head())

from nltk.tokenize import word_tokenize

# Now apply the tokenizer
df['NLTK_Tokens'] = df['Sentence'].apply(word_tokenize)

# Show result
display(df[['Sentence', 'NLTK_Tokens']].head())
```

```python
from transformers import BertTokenizer

# Load the pre-trained BERT tokenizer
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Define a function to apply the tokenizer
def bert_tokenize(text):
    return bert_tokenizer.tokenize(text)

# Apply it to the dataset
df['BERT_Tokens'] = df['Sentence'].apply(bert_tokenize)

# Preview the results
df[['Sentence', 'BERT_Tokens']].head()
```

```python
# Add token count columns
df['Whitespace_Count'] = df['Whitespace_Tokens'].apply(len)
df['Regex_Count'] = df['Regex_Tokens'].apply(len)
df['NLTK_Count'] = df['NLTK_Tokens'].apply(len)
df['BERT_Count'] = df['BERT_Tokens'].apply(len)

# Show comparison
df[['Sentence', 'Whitespace_Count', 'Regex_Count', 'NLTK_Count', 'BERT_Count']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Plot token count distribution for each tokenizer
df_counts = df[['Whitespace_Count', 'Regex_Count', 'NLTK_Count', 'BERT_Count']]
df_counts.plot(kind='box', figsize=(10, 6), title='Token Count Distribution by Tokenizer')

plt.ylabel("Token Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Prepare the average token lengths
avg_tokens = {
    'Whitespace': df['Whitespace_Count'].mean(),
    'Regex': df['Regex_Count'].mean(),
    'NLTK': df['NLTK_Count'].mean(),
    'BERT': df['BERT_Count'].mean()
}

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=list(avg_tokens.keys()), y=list(avg_tokens.values()), palette='muted')
plt.title('Average Number of Tokens per Sentence by Tokenizer')
plt.ylabel('Average Token Count')
plt.xlabel('Tokenizer Type')
plt.show()
```
